```python
ndvi_path = '/Users/tillweiss/Desktop/MODSNOW/data/MODSNOW_irrigated_MarJul_2000_2024/MOD13Q1.061_250m_aid0001.nc'

ds = xr.open_dataset(ndvi_path)

ndvi = ds['_250m_16_days_NDVI']
qa = ds['_250m_16_days_VI_Quality'].astype('uint16')
```

```python
# MODLAND QA (bits 0–1)
modland = qa & 0b11

# VI usefulness (bits 2–5)
usefulness = (qa >> 2) & 0b1111

# Cloud related flags
adjacent_cloud = (qa >> 8) & 1
mixed_cloud    = (qa >> 10) & 1

# Snow / ice
snow_ice = (qa >> 14) & 1

# Land / water (bits 11–13)
landwater = (qa >> 11) & 0b111
```

```python
qa_mask = (
    (modland == 0) &              # good MODLAND QA
    (usefulness <= 2) &            # high usefulness
    (adjacent_cloud == 0) &
    (mixed_cloud == 0) &
    (snow_ice == 0) &
    (landwater == 1)               # land only
)
```

```python
ndvi_clean = ndvi.where(qa_mask)
```

```python
ndvi_clean = ndvi_clean.astype('float32')
```

```python
ndvi_clean.name = 'ndvi_clean'
ndvi_clean.attrs = {
    'description': 'Cleaned NDVI using quality assurance flags',
    'mask': 'qa_mask'
}

ndvi_clean.to_netcdf('NDVI_clean.nc')
```

```python
summer = ndvi_clean.sel(time=ndvi['time.month'].isin([6, 7, 8]))
winter = ndvi_clean.sel(time=ndvi['time.month'].isin([12, 1, 2]))

ndvi_summer_mean = summer.mean(dim='time')
ndvi_winter_mean = winter.mean(dim='time')

# Seasonal amplitude
ndvi_amp = ndvi_summer_mean - ndvi_winter_mean

#Variability filter
ndvi_std = ndvi_clean.std(dim='time')
```

```python
for name, mask in {
    "balanced": ((ndvi_summer_mean>=0.30)&(ndvi_amp>=0.20)&(ndvi_std>=0.04)&(ndvi_std<=0.18)),
    "aggressive": ((ndvi_summer_mean>=0.35)&(ndvi_amp>=0.25)&(ndvi_std>=0.05)&(ndvi_std<=0.16)),
    "wedge": ((ndvi_summer_mean>=0.30)&(ndvi_amp>=0.20)&(ndvi_std>=0.04)&(ndvi_std<=(0.02+0.55*ndvi_amp))),
}.items():
    print(name, float(mask.mean()))
```

```python
# --- 1) flatten to 1D and remove NaNs
amp = ndvi_amp.values.ravel()
std = ndvi_std.values.ravel()

m = np.isfinite(amp) & np.isfinite(std)
amp = amp[m]
std = std[m]

# (optional) ignore negative amplitude / extreme outliers
amp_min, amp_max = 0.0, np.nanpercentile(amp, 99.5)
std_max = np.nanpercentile(std, 99.5)

# --- 2) bin amplitude and compute an "upper envelope" quantile of std per bin
nbins = 40
bins = np.linspace(amp_min, amp_max, nbins + 1)
bin_id = np.digitize(amp, bins) - 1

q = 0.95  # 95th percentile traces the upper edge; try 0.90 or 0.98 too
amp_mid = 0.5 * (bins[:-1] + bins[1:])
std_q = np.full(nbins, np.nan)

for i in range(nbins):
    s = std[bin_id == i]
    if s.size >= 200:  # require enough points for stability; adjust if needed
        std_q[i] = np.nanquantile(s, q)

# keep only bins with values
ok = np.isfinite(std_q)
x = amp_mid[ok]
y = std_q[ok]

slope, intercept = np.polyfit(x, y, 1)

print(f"Envelope (q={q:.2f}): std <= {intercept:.3f} + {slope:.3f} * amp")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(amp[::50], std[::50], s=1, alpha=0.05)  # decimate for speed
plt.plot(x, y, linewidth=2, label=f"{int(q*100)}th pct per bin")
xx = np.linspace(amp_min, amp_max, 200)
plt.plot(xx, intercept + slope*xx, linewidth=2, label="Linear envelope fit")
plt.xlabel("NDVI amplitude")
plt.ylabel("NDVI std")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
ndvi['time'] = pd.to_datetime(ndvi['time'].values)

# --- 2. Load Snow_Percent DataFrame ---  # replace with your file
snow_df['Date'] = pd.to_datetime(snow_df['Date'])

# Align snow data with NDVI times
snow_series = snow_df.set_index('Date')['Snow_Percent'].reindex(ndvi['time'].values)
valid_idx = ~snow_series.isna()
ndvi_filtered = ndvi.isel(time=np.where(valid_idx)[0])
snow_series_filtered = snow_series[valid_idx].values


# --- 3. Prepare NDVI array ---
ndvi_vals = ndvi_filtered.values  # shape (time, lat, lon)
t, lat, lon = ndvi_vals.shape

# Reshape NDVI to 2D: (time, pixels)
ndvi_2d = ndvi_vals.reshape(t, lat*lon)

# --- 4. Standardise (handle NaNs) ---
ndvi_mean = np.nanmean(ndvi_2d, axis=0)
ndvi_std = np.nanstd(ndvi_2d, axis=0)
ndvi_std[ndvi_std == 0] = np.nan  # avoid divide by zero
ndvi_norm = (ndvi_2d - ndvi_mean) / ndvi_std

snow_norm = (snow_series_filtered - np.nanmean(snow_series_filtered)) / np.nanstd(snow_series_filtered)

# --- 5. Pixel-wise correlation ---
# multiply and average along time
corr_flat = np.nanmean(ndvi_norm * snow_norm[:, None], axis=0)

# Reshape back to (lat, lon)
corr_map = corr_flat.reshape(lat, lon)

# --- 6. Wrap as xarray DataArray ---
corr_da = xr.DataArray(
    corr_map,
    coords={'lat': ndvi_filtered.lat, 'lon': ndvi_filtered.lon},
    dims=['lat', 'lon'],
    name='ndvi_snow_corr'
)

# --- 7. Save result ---
corr_da.to_netcdf("ndvi_snow_correlation.nc")

print("Pixel-wise correlation calculation completed.")
```
